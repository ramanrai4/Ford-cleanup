```python
def feature_eng(df1):
        df1 = pd.get_dummies(df1, columns = ['fuel'], drop_first = True, prefix = 'fuel')
        df1 = pd.get_dummies(df1, columns = ['body'], drop_first = True, prefix = 'body')
        df1 = sm.add_constant(df1)
        
        return df1
```
